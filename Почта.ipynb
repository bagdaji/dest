## Очистка данных

- Перевсети слова в нижний регистр
- Удалить все что не является русскими или английскими словами
- удалить стоп слова
- Удалить лишние пространства, если имеются

```python
#df = df.drop("date", axis = 1)
df = df.dropna()
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")
```

```python
def clean_data(df):
    df = df.str.lower()
    df = df.replace(r'[^а-яА-Яa-zA-Z]', ' ', regex = True)
    df = df.str.strip()
    stop_words = stopwords.words("russian")
    df = df.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
    return df
```

```python
df["description"] = clean_data(df["description"])
df["author_company"] = clean_data(df["author_company"])
```

```python
df.head()
```

```python
df["adresat"] = df["adresat"].str.lower()
df["author"] = df["author"].str.lower()
```

### Заключение по первичному обучению

Как видим наша модель методом опорных векторов показала неплохой результат на мультиклассовой классификации. Далее нам необходимо провести Feature Engineering, чтобы привести к наилучшему результату

## Feature Engineering

Идея реконструирования признаков в том, чтобы привести классифкацию к классу важности и сделать ее бинарной:
- Для этого создадим функцию которая будет кодировать наиболее важных представителей WS. А именно `уразов р.н.` и `глушко д.е.`- 1, остальные же отправляются сразу в класс - 0.
- Провести обучение на новой целевой переменной `class_v`

Определяем функцию

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
def class_v(x):
    if x == str("уразов р.н."):
        return 1
    elif x == str("глушко д.е."):
        return 1
    else:
        return 0
```

Создаем новую колонку и заполняем пустыми значениями, после чего используем `lambda`

```python
import numpy as np
df["class_v"] = np.nan
df["class_v"] = df["adresat"].apply(lambda x: class_v(x))
```

Определяем новый `X` и `y`

```python
X = pca_feat_x
y = df["class_v"].values
```

Снова разделяем данные

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, shuffle = True)
```

Тренируем

```python
svc.fit(X_train, y_train)
```

Тестируем на метриках

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, svc.predict(X_test)))
```
